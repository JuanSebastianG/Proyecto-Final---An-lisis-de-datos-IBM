## Tarea 3: Transformación de datos (cualquier herramienta)





- Cambia el nombre de la columna `availability 365` a `days_booked`.
- Convierte todos los nombres de columna a minúsculas y sustituye los espacios en los nombres de columna por un guión bajo "_".
- Elimina el signo de dólares y la coma de las columnas `price` y `service_fee`. Si es necesario, convierte estas dos columnas al tipo de datos adecuado.

```python
## Cambia el nombre de la columna.
lista_limpia.rename(columns={'availability 365': 'days_booked'}, inplace=True)
```

```python
## Convierte todos los nombres de columna a minúsculas y sustituye los espacios por un guión bajo "_".
lista_limpia.columns = lista_limpia.columns.str.lower().str.replace(' ', '_')
```

```python
## Elimina el signo de dólares y la coma de las columnas. Si es necesario, convierte estas dos columnas al tipo de datos adecuado.
lista_limpia['price'] = lista_limpia['price'].replace({'\$': '', ',': ''}, regex=True)
lista_limpia['service_fee'] = lista_limpia['service_fee'].replace({'\$': '', ',': ''}, regex=True)
lista_limpia['price'] = pd.to_numeric(lista_limpia['price'], errors='coerce')
lista_limpia['service_fee'] = pd.to_numeric(lista_limpia['service_fee'], errors='coerce')
```

### Tarea 4: Análisis exploratorio de datos (cualquier herramienta)



- Enumera los tipos de habitaciones disponibles en el <i>dataset</i>.
- ¿Qué tipo de habitación tiene la política de cancelación más estricta?
- Enumera el precio medio por barrio y señala cuál es el conjunto de barrios más caro para alquilar.



Si utilizas Python para este ejercicio, incluye el código que hayas utilizado en las celdas siguientes. Si utilizas cualquier otra herramienta, incluye capturas de pantalla de tu trabajo.

```python
## Enumera los tipos de habitaciones disponibles en Airbnb.
tipo_de_habitaciones = lista_limpia ['room_type'].value_counts()
print("Tipos de habitaciones disponibles en Airbnb:")
print(tipo_de_habitaciones)
```

```python
## ¿Qué tipo de habitación se adhiere a una política de cancelación más estricta?
politica_cancelacion_estricta = lista_limpia.groupby('room_type')['cancellation_policy'].value_counts().groupby(level=0).head(1)
print("Tipos de habitación con políticas de cancelación estricta: ")
print(politica_cancelacion_estricta)
```

```python
## Enumera los precios por barrio y menciona también cuál es el grupo de barrios con alquileres más caros.
barrio_precio_promedio = lista_limpia.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
print("Precio medio por grupo de vecinos:")
print(barrio_precio_promedio)

barrio_mas_caro = barrio_precio_promedio.idxmax()
print("\nEl barrio más caro para alquilar es:", barrio_mas_caro)

### Prueba de gráfico
plt.figure(figsize=(12, 8))
barrio_precio_promedio.sort_values().plot(kind='barh', color='teal')
plt.title('Precio Promedio por Grupo Vecinal')
plt.xlabel('Precio Promedio')
plt.ylabel('Grupo Vecinal')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(x='neighbourhood_group', y='price', data=lista_limpia, palette='viridis')
plt.title('Distribución de Precios por Vecinos')
plt.xlabel('Grupo de Vecinos')
plt.ylabel('Precio')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(10, 6))
sns.stripplot(x='neighbourhood_group', y='price', data=lista_limpia, jitter=True, color='purple', alpha=0.7)
plt.title('Distribución de Precios por Vecinos')
plt.xlabel('Grupo de Vecinos')
plt.ylabel('Precio')
plt.xticks(rotation=90)
plt.show()
```

```python
## Enumere los 10 barrios más caros por orden creciente de precio con la ayuda de un gráfico de barras horizontales. ¿Cuál es el barrio más barato?
barrio_por_precio = lista_limpia.groupby('neighbourhood')['price'].mean().sort_values().head(10)
plt.figure(figsize=(10, 6))
barrio_por_precio.plot(kind='bar', color='skyblue')
plt.title('Top 10 barrios principales por precio (Desde el más económico a más costoso)')
plt.xlabel('Vecindario')
plt.ylabel('Precio promedio')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Enumere los barrios que ofrecen alquileres a corto plazo de menos de 10 días. Ilustrar con un gráfico de barras
corto_plazo = lista_limpia[lista_limpia['minimum_nights'] <= 10]['neighbourhood'].value_counts().head(10)
plt.figure(figsize=(10, 6))
corto_plazo.plot(kind='bar', color='salmon')
plt.title('Barrios que ofrecen alquileres a corto plazo (<= 10 días)')
plt.xlabel('Vecindario')
plt.ylabel('Contar')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Enumere los precios con respecto al tipo de habitación utilizando un gráfico de barras y exponga también sus inferencias.
plt.figure(figsize=(8, 6))
sns.barplot(x='room_type', y='price', data=lista_limpia, ci=None)
plt.title('Precio según tipo de habitación')
plt.xlabel('Tipo de habitación')
plt.ylabel('Precio')
plt.show()
```

```python
# Cree un gráfico circular que muestre la distribución de los días reservados para cada grupo de barrios.
dias_de_reserva = lista_limpia.groupby('neighbourhood_group')['minimum_nights'].sum()
plt.figure(figsize=(8, 8))
pie_chart = dias_de_reserva.plot(kind='pie', autopct='%1.1f%%', startangle=140, labeldistance=1.1)
plt.title('Distribución de días reservados para cada grupo vecinal')
plt.ylabel('')
plt.show()
```

Si utiliza Python para este ejercicio, por favor incluya el código en las celdas de abajo. Si utiliza cualquier otra herramienta, por favor incluya capturas de pantalla de su trabajo.

```python
# Efecto del número de tasa de revisión en el precio
plt.figure(figsize=(8, 6))
plt.boxplot([lista_limpia[lista_limpia['review_rate_number'] == i]['price'].dropna() for i in lista_limpia['review_rate_number'].unique()], 
            labels=lista_limpia['review_rate_number'].unique())
plt.title('Efecto del número de tasa de revisión sobre el precio')
plt.xlabel('Número de tarifa de revisión')
plt.ylabel('Precio')
plt.grid(True)
plt.show()
```

```python
# Efecto de la identidad del host verificada en el precio

plt.figure(figsize=(8, 6))
plt.boxplot([lista_limpia[lista_limpia['host_identity_verified'] == i]['price'].dropna() for i in lista_limpia['host_identity_verified'].unique()], 
            labels = lista_limpia['host_identity_verified'].unique())
plt.title('Efecto de la identidad del anfitrión verificada en precio')
plt.xlabel('Identidad del anfitrión verificada')
plt.ylabel('Precio')
plt.grid(True)
plt.show()
```

```python
# Cree un diagrama de barras para obtener el número máximo de reseñas por grupo vecinal

max_reseñas_por_grupo = lista_limpia.groupby('neighbourhood_group')['number_of_reviews'].max()
faltan_reseñas_maximas = lista_limpia[lista_limpia['neighbourhood_group'].isna()]['number_of_reviews'].max()

data = pd.DataFrame({
    'Grupo Vecinal': max_reseñas_por_grupo.index,
    'Máximo Número de Reseñas': max_reseñas_por_grupo.values
})

missing_neighborhood_data = pd.DataFrame({
    'Grupo Vecinal': 'Vecindarios Faltantes',
    'Máximo Número de Reseñas': faltan_reseñas_maximas
}, index=[0])

data = pd.concat([data, missing_neighborhood_data], ignore_index=True)
display(data)
plt.figure(figsize=(10, 6))
plt.bar(data['Grupo Vecinal'], data['Máximo Número de Reseñas'], color='skyblue')
plt.xlabel('Grupo Vecinal')
plt.ylabel('Máximo Número de Reseñas')
plt.title('Máximo Número de Reseñas por Grupo Vecinal')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
